## 変数について

変数は、例えるなら『箱』のことです 


変数があることで文字や数字などのデータを変数に入れておくことができます。

変数にデータを入れることを**『代入』**といい、

取り出すことを**『参照』**と言います

変数には名前をつけることができ、**変数名**と言います

それでは実際のコードで確認していきましょう！

下記コードは変数`num`に整数の1を代入し、その変数の中身を`print`を使って表示させるプログラムです

変数へ代入するときは、イコール`=`を使います

```python
num = 1　#変数numに1を代入
print(num)　#numを出力
```

## 変数名のルール

### ■ルール１_使用できる文字

変数名には、**アルファベット**、**数字**、**アンダースコア**(`_`)が使えます

変数名がつけられるか、あるいはエラーになるか試してみましょう

`num`のあとに01の数字をつける変数を作ってみます

```python
num = 1
num01 = 2
num_01 = 3

print(num)
print(num01)
print(num_01)
```

### ■ルール2_使用できない文字、パターン

変数名は、**数字**から始めることができません

また、**アンダースコア**(`_`)以外の記号を使うことはできません。

```python
num$01 = 4
num-01 = 5
01num = 6

print(num$01)
print(num-01)
print(01num)
```

↑変数名が赤い文字になっています。エラーです。

### ■ルール3_区別について

変数名の**大文字**と**小文字**は区別されます

すべて大文字の`NUM`、

最初だけ大文字の`Num`を作って変数を代入させてみましょう

```python
NUM = 1
Num = 2

print(NUM)
print(Num)
```

### ■ルール4_変数名に使用できない単語について

予約語は変数名にすることができません。

予約語とは、『`return`』、『`class`』、『`for`』、『`while`』など、

プログラミング言語ですでに役割が決まっている単語のことをいいます

実際に`return`という変数を作ってみます

```python
return = 10

print(return)
```

実行するとエラーになることがわかります

## データ型について

データ型とは、変数に入れるデータの種類のことです。

今回は**数値型**、**文字列型**、**ブール型**について説明します

### ■数値型について

数値型には、整数の『**`int型`**』、小数点の『**`float型`**』があります。

`num01`には整数の123を代入し、`num02`には小数点の1.23を代入します

```python
num01 = 123
num02 = 1.23

print(num01)
print(num02)
```

Pythonのデータは**`type(タイプ)`**を使用して確認することができます

`**type()**`の丸括弧の中に変数名を入れて`print`で実行してみましょう

```python
num01 = 123
num02 = 1.23

print(type(num01))
print(type(num02))
```

### ■文字列型について

文字列型は『**`String型`**』ともいいます

**`string_a`**という変数に文字列を代入してみましょう

文字列はダブルクォーテーション`" "`、もしくはシングルクォーテーション`' ' `でくくります

```python
string_a = "Hello Word!"

print(string_a)
print(type(string_a))
```

### ■ブール型について

ブール型は、**`Boolean型(ブーリアン型)`**ともいいます

ブール型は、**`True`**また**`Flase`**の２つのうち、どちらか1つを持つ型です

そもそも`True`と`False`とはなんでしょうか？

例えば10と1では10の方が大きいです。 <br>
つまり『**` 10 > 1`**』は正しいです。

これを変数に代入すると**`True`**が入ります。

逆に『**` 10 < 1`**』は誤りです。 <br>
これを変数に代入すると**`False`**が入ります

じっさいにコードで確認してみましょう！

```python
a = 10 #10を代入
b = 1 #1を代入

bool01 = (a > b)

print(bool01)
print(type(bool01))
```
